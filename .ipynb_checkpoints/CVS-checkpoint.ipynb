```python
#data preprocessing
#training data set
training_set = cvs_df[["High", "Low", "Open", "Close","Volume","Adj Close"]]
training_set = pd.DataFrame(training_set)
```

```python
#normalize the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform (training_set)
scaled_data
```

```python
#spilt data into training 60% of the data and then testing 40% 
train_size = int(len(cvs_df) * 0.60)
test_size = len(cvs_df) - train_size
train, test = cvs_df[0:train_size],cvs_df[train_size:len(cvs_df)]
```

```python
#values for range
count_cvs = len(cvs_df)
count_cvs
```

```python
# training data from Day 1 to Day 60 and make predictions from 61 forward 
# then train data from Day 2 to Day 61 and make predictions from 62 forward
# range value 7642
def create_dataset(cvs_df, lookback =1):
    x_train = []
    y_train = []
    #append the last 60 values
    for i in range (60,7642):
        x_train.append(scaled_data[i-60:i, 0])
        y_train.append(scaled_data[i, 0])

    return np.array(x_train), np.array(y_train)
```

```python
#prep for reshaping 
look_back = 1
x_train, y_train = create_dataset(train, look_back)
test_x, test_y = create_dataset(test, look_back)
```

```python
#reshaping the trained data into samples,timesets and features
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
test_x = np.reshape(test_x, (test_x.shape[0], test_x.shape[1], 1))
```

```python
#feature extraction ("open","high", "low", "close", adjusted close","volume", "dividend amount", "split coefficent")
#RNN(recurrent neural networks) initiation #LSTM model
from keras.layers import Dropout 
model = Sequential()
#input layer, first LSTM layer 
model.add(LSTM( 50, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(Dropout(0.2))
#second LSTM layer
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
#third LSTM layer
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
#output layer single value prediction
model.add(Dense(1))
```

```python
#Adam optimizer used to calculate individual learning rates for each parameter by learning from old values
#mix of AdaGrad+RMSprop
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#begin training
#epoch= # of passes through the dataset
model.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
#make predictions
train_predictions = model.predict(x_train)
test_predictions = model.predict(test_x)
```

```python
#evaluating the training set to determine error
#getting the root mean squared error
rmse = np.sqrt(np.mean(y_train[0], train_predictions[:,0]))
rmse
```
