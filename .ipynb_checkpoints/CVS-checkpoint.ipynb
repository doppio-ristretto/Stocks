```python
#data preprocessing
#dataset by features
training_set = cvs_df[["High", "Low", "Open", "Close","Volume","Adj Close"]]
training_set = pd.DataFrame(training_set)
```

```python
training_set.shape
```

```python
#normalize the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform (training_set)
scaled_data
```

```python
#spilt data into training 60% of the data and then testing 40% 
#used to verify  x_train amount of data points is correct
train_size = int(len(scaled_data) * 0.60)
test_size = len(scaled_data) - train_size
train, test = scaled_data[0:train_size],scaled_data[train_size:len(scaled_data)]
```

```python
train.shape
```

```python
test.shape
```

```python
# separating dataset into x_train and y_train
#list
x_train = []
y_train = []
#append the last 60 timestamps
for i in range(int(len(scaled_data) * 0.40),len(training_set)):
    x_train.append(scaled_data[i-60:i, 0])
    y_train.append(scaled_data[i,0])
    
```

```python
#convert the x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
#reshaping the trained data into samples,timesets and features
#confirming 60% for training set (rows, columns)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape
```

```python
#feature extraction ("open","high", "low", "close", adjusted close","volume")
#RNN(recurrent neural networks) initiation #LSTM model
from keras.layers import Dropout 
model = Sequential()
#input layer, first LSTM layer 
#multiple layers to boost learning of the model
model.add(LSTM( 50, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(Dropout(0.2))#regularization techinque used to boost accuracy by 2% 
#second LSTM layer
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
#third LSTM layer
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
#output layer single value prediction
model.add(Dense(units = 1))
```

```python
#hyperparameter optimization
#Adam optimizer used to calculate individual learning rates for each parameter by learning from old values
#mix of AdaGrad+RMSprop
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#begin training
#epoch= # of passes through the dataset
model.fit(x_train, y_train, batch_size=32, epochs=100)
```
