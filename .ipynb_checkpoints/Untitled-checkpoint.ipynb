```python
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#retrieving historical data
cvs_df = web.DataReader('CVS', data_source='yahoo', start='2012-01-01', end = '2020-4-30')
#displaying only for reference
cvs_df.head(20)
```

```python
#retrieving the number of rows
cvs_df.shape
```

```python
#visualizing historical data
plt.figure(figsize=(16, 8))
plt.title('Close Pricing History')
plt.plot(cvs_df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price')
plt.show
```

```python
#filter data to show only closing price
data = cvs_df.filter(['Close'])

#convert data into a numpy array for training purposes
dataset = data.values

#count the number of rows to train the model 
training_data_len = math.ceil( len(dataset) * .8 )

training_data_len
```

```python
#scale the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform (dataset)

scaled_data
    
```

```python
#creating a scaled data set
train_data = scaled_data[0:training_data_len , :]
#spilt data into x_train and y_training sets
x_train = []
y_train = []
#append the last 60 values
for i in range (60, len(train_data)):
  x_train.append(train_data[i-60:i, 0])
  y_train.append(train_data[i, 0])
  if i<=60:
    print(x_train)
    print(y_train)
    print()
```

```python
#convert the x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
#reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape
```

```python
#LSTM model
model = Sequential()
model.add(LSTM( 50, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
#complie the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#begin training
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
test_data = scaled_data[training_data_len - 60: , :]
x_test = []
y_test = dataset[training_data_len: , :]
for i in range(60, len(test_data)):
  x_test.append(test_data[ i-60: i, 0])
```

```python
x_test = np.array(x_test)
```

```python
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
```

```python
#getting the root mean squared error
rmse = np.sqrt( np.mean(predictions - y_test)**2 )
rmse
```

```python
# plot the data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
plt.figure(figsize=(16,8))
plt.title ('Model')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Predicted Closing Price', fontsize = 18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val','Predictions'], loc='lower right')
plt.show()
```
